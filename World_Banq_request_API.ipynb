### Request an API on World Bank

```python
with open ('countries.pickle','rb') as fp: #rb : read bite
    countries = pickle.load(fp)
```

```python
def get_indicator(country_code,indicator):
    head = 'http://api.worldbank.org/v2/country/'
    middle = '/indicator/'
    tail = '?format=json'
    url = head + country_code + middle + indicator + tail
    r = requests.get(url)
    try:
        data = r.json()
        for i in data[1]:
            if i['value'] is None:
                continue
            else:
                return i['value']
    except KeyError:
        return None
    except IndexError:
        return None
```

```python
def get_value_on_WB(header, indicator):
    for i,v in df['Country (according to CN Gov):'].iteritems():
        country_name = v.split('(')[0].strip()
        country_code = countries[country_name]
        value = get_indicator(country_code, indicator)
        if value is not None:
            df.at[i, header] = value
        print(country_name,'done...')
        time.sleep(0.5+random.random())
    df[['Country (according to CN Gov):', header]]
```

```python
def get_value_for_each_column(dictionnaire):
    for i in dictionnaire:
        get_value_on_WB(i, dictionnaire[i])
        
print (get_value_for_each_column(indicator_list))
```
